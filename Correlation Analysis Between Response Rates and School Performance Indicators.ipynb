```python
columns_to_convert = [
    'Parent Response Rate',
    'Teacher Response Rate',
    'Student Response Rate',
    'Total Safety and Respect Score',
    'Total Communication Score',
    'Total Engagement Score',
    'Total Academic Expectations Score'
]

for col in columns_to_convert:
    df[col] = df[col].astype(str)                # Ensure values are treated as strings
    df[col] = df[col].str.replace('%', '', regex=False)  # Remove the percent sign
    df[col] = df[col].str.strip()                # Trim any leading/trailing whitespace
    df[col] = pd.to_numeric(df[col], errors='coerce')    # Convert to numeric (floats), set invalid values to NaN

# Check data types after conversion
print(df[columns_to_convert].dtypes)
```

```python
import pandas as pd

# List of indicators to analyze
columns_to_check = ['Parent Response Rate', 'Teacher Response Rate', 'Student Response Rate', 'Performance']

# List to store top school info
top_schools = []

# Loop through each column and get the top school
for col in columns_to_check:
    max_index = df[col].idxmax()
    school_type = df.loc[max_index, 'School Type'] if 'School Type' in df.columns else 'Unknown'
    score = df.loc[max_index, col]
    top_schools.append({
        'Indicator': col,
        'Top School Type': school_type,
        'Top Value (%)': f"{round(score * 100, 1)}%"
    })

# Convert to DataFrame
top_df = pd.DataFrame(top_schools)

# Style the DataFrame with colors
styled_table = top_df.style.set_caption("🏆 Top Performing School Type per Indicator")\
    .set_table_styles([{
        'selector': 'th',
        'props': [('background-color', '#BAE1FF'), ('color', '#000'), ('font-weight', 'bold')]
    }])\
    .applymap(lambda val: 'background-color: #BAFFC9; font-weight: bold;' if '%' in str(val) else '', subset=['Top Value (%)'])\
    .applymap(lambda val: 'background-color: #FFDFBA;' if isinstance(val, str) else '', subset=['Indicator', 'Top School Type'])

# Display the styled table
styled_table

```

```python

data = {
    'School Type': ['Elementary School', 'Elementary School', 'Middle School', 'High School', 'Middle School'],
    'Parent Response Rate': [61, 92, 39, 70, 50],
    'Teacher Response Rate': [92, 94, 58, 80, 60],
    'Student Response Rate': [None, None, 96, 85, 90]
}

df = pd.DataFrame(data)

for col in ['Parent Response Rate', 'Teacher Response Rate', 'Student Response Rate']:
    df[col] = pd.to_numeric(df[col], errors='coerce') / 100


response_means = df.groupby('School Type')[
    ['Parent Response Rate', 'Teacher Response Rate', 'Student Response Rate']
].mean().round(2)


pastel_colors = ['#FFB3BA', '#FFDFBA', '#FFFFBA']


ax = response_means.plot(kind='bar', figsize=(14, 8), color=pastel_colors)

plt.title('Average Response Rates by School Type', fontsize=16, fontweight='bold', color='#333333')
plt.xlabel('School Type', fontsize=14, color='#333333')
plt.ylabel('Average Response Rate', fontsize=14, color='#333333')
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.ylim(0, 1)
plt.grid(axis='y', linestyle='--', alpha=0.6)


ax.legend(title='Response Type', fontsize=12, title_fontsize=13,
          loc='upper center', bbox_to_anchor=(0.5, -0.45), ncol=3)

plt.tight_layout(rect=[0, 0.05, 1, 1])


for p in ax.patches:
    height = p.get_height()
    if pd.notna(height):
        ax.annotate(f'{height:.0%}',
                    (p.get_x() + p.get_width() / 2, height),
                    ha='center', va='bottom',
                    fontsize=11, fontweight='bold',
                    color='#333333')

plt.show()
```
